Similar al anterior, pero puedes limitar el número de intercambios conversacionales almacenados en la memoria. Por ejemplo, puedes configurarlo para que solo recuerde las últimas 3 preguntas y respuestas de la conversación.

```python
from langchain_anthropic import ChatAnthropic
```

```python
from google.colab import userdata
anthropic_api_key = userdata.get('ANTHROPIC_API_KEY')
```

```python
llm = ChatAnthropic(model='claude-3-opus-20240229', temperature=0, api_key=anthropic_api_key)
```

```python
from langchain.memory import ConversationBufferWindowMemory
window_memory = ConversationBufferWindowMemory(k=2)
```

```python
from langchain.chains import ConversationChain
conversation = ConversationChain(
    llm=llm,
    memory = window_memory,
    verbose=True
)
```

```python
conversation.invoke(input="Hola, mi nombre es Juan. ¿Te gusta el cine?")
```

```python
conversation.invoke(input="¿Cuáles son tus películas favoritas?")
```

```python
conversation.invoke(input="¿Te gusta el cine argentino? ¿Cuáles son tus películas favoritas?")
```

```python
conversation.invoke(input="¿Recuerdas mi nombre?")
```
